```python
import pandas as pd
import numpy as np
import re
```

```python
df=pd.read_csv("/content/nlp_preprocessing_dataset.csv")
```

```python
df
```

```python
def remove_new_line(text):
  pattern="\n"
  clean_text=re.sub(pattern,"",text)
  return clean_text
```

```python
df["text"]=df["text"].apply(remove_new_line)
```

```python
df
```

```python
def remove_html_tags(text):
  pattern="<.*?>"
  clean_text=re.sub(pattern,"",text)
  return clean_text
```

```python
df["text"]=df["text"].apply(remove_html_tags)
```

```python
df
```

```python
import string
```

```python
exclude=list(string.punctuation)
```

```python
exclude
```

```python
def remove_punctuation(text):
  pattern="[" + re.escape("".join(exclude)) + "]"
  clean_text=re.sub(pattern,"",text)
  return clean_text
```

```python
df["text"]=df["text"].apply(remove_punctuation)
```

```python
df
```
