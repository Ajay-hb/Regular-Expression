-regular expression we are using to match the pattern

-regular expression have there own characters

-we always write a regular expression in verbose form.

-regular espression alwys scan from left to right.

syntax:

    import re

functions in regular expression:

1.re.match()

2.re.findall()

3.re.sub()

4.re.search()

5.re.compile

characters in ReGEX:
1.  .(dot)--->it will match a single character

    .. --->2characters
	...--->3 characters
2. *---->match 0 or more chracters

3. +--->match 1 or more chracters

4. ?--->0 or 1 character

5. [a]--->matches a

   [ab]--->a or b

   [abc]--->a or b or c                          Meta Characters

   [a-zA-Z0-9]

```python
string="mayurkulkarni6667@gmail.com"
pattern=r"[a-z0-9]+\@[a-z]+\.[a-z]+"
print(re.findall(pattern,string))
```

```python
str1="mayur@"
pattern=r"[a-z]+\@"
print(re.findall(pattern,str1))
```

```python
str2="mayur133@#gmail.co.in"
pattern=r"[a-z0-9]+\@\#[a-z]+\.[a-z]+\.[a-z]+"
print(re.findall(pattern,str2))
```

```python
str3="I have a 4 apples and 3 oranges"
pattern=r"\d"
print(re.findall(pattern,str3))
```

```python
str4="I born in 28 Nov 1996"
pattern=r'\d+'
print(re.findall(pattern,str4))
```
